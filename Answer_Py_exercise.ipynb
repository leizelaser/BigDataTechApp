```python
Fibonacci = [1.0,1.0]
for i in range(0, 98):
    Fibonacci.append(Fibonacci[-1] + Fibonacci[-2])
Fibonacci_rate = [Fibonacci[i]/Fibonacci[i+1] for i in range(0,99)]
```

```python
p = numpy.polyfit(numpy.linspace(1,100,100), numpy.log(numpy.array(Fibonacci)), 1)
FitFibonacci = numpy.exp(numpy.linspace(1,100,100) * p[0] + p[1])
DiffFibonacci = Fibonacci - FitFibonacci
```

```python
fig, ax = plt.subplots()
ax.set_yscale("log")
ax.plot(numpy.linspace(1,100,100), Fibonacci, "b+", label="Fibonacci Series")
ax.plot(numpy.linspace(1,100,100), FitFibonacci, "r--", label="Fibonacci Fitting")
plt.legend()
plt.show()
```
